```python
import glob
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pylab as plt
import chainer
import chainer.functions as F
import chainer.links as L
from chainer.training import extensions
from PIL import Image
```

```python
label_files = glob.glob('./data/*_L.png')
files = glob.glob('./data/*.png')
image_files = list(set(files)-set(label_files))

label_files.sort()
image_files.sort()

len(image_files), len(label_files)
```

```python
for i, (image, label) in enumerate(zip(image_files, label_files)):
    
    print(image, label)
    
    if (i+1) == 10:
        break
```

```python
indices = np.random.randint(len(image_files), size=3)

for i in indices:
    
    fig, axs = plt.subplots(ncols=2, figsize=(10,5))
    
    axs[0].imshow(Image.open(image_files[i]))
    axs[0].axis('off')
    axs[1].imshow(Image.open(label_files[i]))
    axs[1].axis('off')
    
    plt.show()
```
