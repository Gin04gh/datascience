```python
import json
import urllib.request
import numpy as np
import pandas as pd
```

```python
f = open('.appid')
j = json.loads(f.read())
appid = j['appid']
f.close()
```

```python
uri = 'http://api.e-stat.go.jp/rest/2.1/app/getSimpleStatsData?cdCat01=011100010'\
    '&cdTime=2018000101'\
    '&cdArea=01003%2C02003%2C03003%2C04003%2C05003%2C06003%2C07003%2C08003%2C09003%2C10003%2C11003%2C12003%2C13003%2C14003%2C15003%2C16003%2C17003%2C18003%2C19003%2C20003%2C21003%2C22003%2C23003%2C24003%2C25003%2C26003%2C27003%2C28003%2C29003%2C30003%2C31003%2C32003%2C33003%2C34003%2C35003%2C36003%2C37003%2C38003%2C39003%2C40004%2C41003%2C42003%2C43003%2C44003%2C45003%2C46003%2C47003%2C14004%2C14150%2C22004%2C27004%2C40003'\
    '&cdCat02=03'\
    '&appId={}'\
    '&lang=J'\
    '&statsDataId=0003103532'\
    '&metaGetFlg=Y'\
    '&cntGetFlg=N'\
    '&sectionHeaderFlg=1'\
    .format(appid)
result = urllib.request.urlopen(uri).read().decode('utf8')
```

```python
result.split(',')[:70]
```

```python
df_payment_sake = pd.DataFrame(columns=['jiscode', 'pref_name', 'payment_sake'])

result_rows = result.splitlines()[28:]

for i, row in enumerate(result_rows):
    
    row_arr = row.split(',')
    jiscode, pref_name = row_arr[7].replace('"', '').split(' ')
    payment_sake = row_arr[-1].replace('"', '')
    
    s = pd.Series([jiscode, pref_name, payment_sake], index=df_payment_sake.columns, name=str(i))
    df_payment_sake = df_payment_sake.append(s)
    
df_payment_sake
```
