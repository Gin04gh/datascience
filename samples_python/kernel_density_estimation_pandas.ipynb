* kernel density estimation by pandas.

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
display(df.head())

target_col = 'petal length (cm)'
df[target_col].plot(kind='hist', bins=20, figsize=(10, 5), color='gray', alpha=0.5) # histogram
df[target_col].plot(kind='kde', figsize=(10,5), secondary_y=True) # density estimate
plt.show()
```
