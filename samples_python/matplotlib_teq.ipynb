```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pylab as plt
```

```python
# 列でカラースケールを対応させた表

from matplotlib.colors import LinearSegmentedColormap

def generate_cmap(colors):
    
    values = range(len(colors))
    vmax = np.ceil(np.max(values))
    color_list = []
    for v, c in zip(values, colors):
        color_list.append((v/vmax,c))
    
    return LinearSegmentedColormap.from_list('custom_cmap', color_list)

cm = generate_cmap(['white', 'gray'])

n_row = 10
names = list(map(lambda x: 'Name'+str(x), list(range(1, n_row+1))))
cols = ['c1', 'c2', 'c3']
data = np.empty((len(cols), len(names)))
for i in range(len(cols)):
    for j in range(len(names)):
        data[i, j] = np.random.rand()*(i+1)
        
fig, axs = plt.subplots(ncols=3, figsize=(4, 6))
plt.subplots_adjust(wspace=0)

for i, c in enumerate(cols):
    
    axs[i].imshow(data[i, :].reshape(len(names), 1), cmap=cm, interpolation='nearest', aspect='auto')
    
    axs[i].xaxis.tick_top()
    axs[i].set_xticks([0])
    axs[i].set_yticks(list(range(len(names))))
    axs[i].set_xticklabels([c])
    axs[i].tick_params(bottom=False, left=False, right=False, top=False)
    
    if i == 0:
        axs[i].set_yticklabels(names)
        axs[i].set_ylabel('Name')
    else:
        axs[i].set_yticklabels([])
    
    for j in range(len(names)):
        axs[i].text(0, j, str(round(data[i, j], 3)), ha='center', va='center', color='black')
        
plt.show()
```
