* 重回帰分析

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import linear_model

iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
df = df.sample(frac=1)
display(df.head())

train_x = df.loc[:, ['sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
train_y = df.loc[:, 'sepal length (cm)']
model = linear_model.LinearRegression()
model.fit(train_x, train_y)

res = pd.DataFrame(data=[model.coef_], columns=train_x.columns, index=['偏回帰係数'])
display(res)
print('切片: {}'.format(model.intercept_)) 
print('自由度調整済み決定係数: {}'.format(model.score(train_x, train_y)))
```
