**1. Plotlyについて**

---

**（はじめに）Pythonで使えるグラフ作成ライブラリ**

基本的な静止グラフ

* **Matplotlib**: Pythonでのグラフ作成では標準的なライブラリ.
* **Pandas**: 描画機能. 内部で Matplotlib を実行.
* **Seaborn**: Matplotlib のラッパーライブラリ.

Pythonでの静止グラフ作成に関しては, 基本的には上記のライブラリが使われる.

インタラクティブなグラフ

* **[Plotly](https://plot.ly/feed/)**: 今回紹介するライブラリ. インタラクティブなグラフを作成する. D3.js で作成.
* **[Bokeh](https://bokeh.pydata.org/en/latest/)**: インタラクティブなグラフを作成する. D3.js で作成.
* **[py_d3](https://github.com/ResidentMario/py_d3)**: Jupyter notebook 上で D3.js を実行できるようにする. 
* **[mpld3](http://mpld3.github.io/)**: Matplotlib のグラフを D3.js に変換するライブラリ.
* **[VisPy](http://vispy.org/)**: OpenGLを使って実行. 科学計算のビジュアライズに使われる.

Pythonでのインタラクティブなグラフに関しては, Bokeh か Plotly のどちらかが選択される印象.

**Plotly**

* 公式: https://plot.ly/feed/
* インタラクティブなグラフを作成できるオープンソースライブラリ
* Python, R, JavaScript, MATLAB, Ruby, Go, Julia ...など様々な言語から実行できる
* グラフの作成は D3.js をベースとしている
    * 作成できるグラフ: Scatter, Bar, Line, Pie, Histogram, Box, Heatmap, Candlestick, Map, 3D Scatter, 3D Surface, 3D etc...
* 以前は3Dプロットが有償であったが, 現在は無償で利用可能
    * Bokehには, 3Dプロットがない
* オンライン・オフラインで実行可能
    * 以前はオンラインのみであったが, 現在はオフラインで利用可能に
    * オンライン
        * Plotlyのサイトでアカウント登録が必要
            * Log in / Sign up: https://plot.ly/accounts/login/?next=%2Fsettings
        * グラフをアップロードできる（制限あり, 有償で拡張）
        * アカウントページでもグラフ作成, 共有が可能
        * アップロードされたグラフは基本的に public だが, 有償で private にもできる
    * オフライン
        * アカウントの登録は不要（ライブラリをインストールして普通に利用するだけ）

---

**初期化**

---

**オンラインで利用する場合**

---

```python
import plotly
plotly.tools.set_credentials_file(username='<user name>', api_key='<api key>')
```

---

上記を実行すると, 以下のファイルが作成される.

```
$ cat ~/.plotly/.credentials
{
    "proxy_username": "",
    "stream_ids": [],
    "api_key": "<api key>",
    "proxy_password": "",
    "username": "<user name>"
}
```

この状態で, Plotlyを実行する.

---

```python
import numpy as np

x = np.random.normal(loc=0, scale=1, size=300)
y = np.random.normal(loc=0, scale=1, size=300)

trace = plotly.graph_objs.Scatter(x=x, y=y, mode='markers')
data = [trace]
layout = plotly.graph_objs.Layout(title='Sample')
figure = plotly.graph_objs.Figure(data=data, layout=layout)
plotly.plotly.iplot(figure)
```
